```python
#dropping unnecessary columns
cols_to_drop = [
    'resultId', 'raceId', 'driverRef', 'constructorRef', 'circuitRef',
    'name_y', 'lat', 'lng', 'alt', 'dob', 'surname', 'forename',
    'nationality_x', 'nationality_y'
]

df = df.drop(columns=cols_to_drop)
```

```python
import pandas as pd
import numpy as np

def lap_time_to_seconds(lap_time):
    if lap_time == '\\N' or lap_time == 0:
        return 0.0
    try:
        # Split minutes and seconds
        if ':' in str(lap_time):
            minutes, seconds = lap_time.split(':')
            return float(minutes) * 60 + float(seconds)
        else:
            return float(lap_time)
    except:
        return 0.0

cols = ['fastestLap', 'milliseconds', 'rank', 'fastestLapTime', 'fastestLapSpeed']

for col in cols:
    # Create flag for missing values
    df[f'{col}_missing'] = df[col].apply(lambda x: 1 if x == '\\N' else 0)

    if col == 'fastestLapTime':
        # Convert lap time to total seconds
        df[col] = df[col].apply(lap_time_to_seconds)
    else:
        # Replace '\N' with 0 and convert to float
        df[col] = df[col].replace('\\N', 0).astype(float)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['year_from_date'] = df['date'].dt.year
```

```python
#now removing outliers

numeric_cols = ['year', 'round', 'grid', 'positionOrder', 'points', 'laps',
                'milliseconds', 'fastestLap', 'rank', 'fastestLapTime', 'fastestLapSpeed']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numeric_cols])
```

```python
z_scores = np.abs(scaled_data)  # absolute value of Z
```

```python
threshold = 3
df_no_outliers = df[(z_scores < threshold).all(axis=1)]
```

```python
print("Original rows:", len(df))
print("Rows after removing outliers:", len(df_no_outliers))
```
